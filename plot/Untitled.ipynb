```python
with open("./主线/序章/raw/序章(上).txt") as f:
    lines = f.readlines()
    for line in lines:
        if "[name=" in line:
            name = line.split("name=")[1].split("]")[0]
            if name == "":
                name = "旁白或者未知"
            if len(name) <= 10:
                while len(name) < 10:
                    name = name + " "
            text = line.split("]")[1]
            print(name, text)
```
